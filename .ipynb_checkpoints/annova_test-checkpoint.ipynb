```python
import seaborn as sns
import pandas as pd
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df = df[['age','pclass']].dropna()
```

```python
df['pclass'].unique()
```

```python
class1 = df[ df['pclass'] == 1 ]['age']
print(class1)
class2 = df[ df['pclass'] == 1 ]['age']
print(class2)
class3 = df[ df['pclass'] == 1 ]['age']
print(class3)
```

```python
f_stats, p_value = f_oneway(class1,class2,class3)
```

```python
print(f_stats)
print(p_value)
```

```python
alpha = 0.05

if p_value<alpha:
    print('rejected')
else:
    print('accepted')
```
