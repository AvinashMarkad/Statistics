### chi-square test

```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
df = sns.load_dataset('titanic')
df
```

```python
contengency_table = pd.crosstab(df['sex'],df['survived'])
contengency_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(contengency_table)
```

```python
print(chi2)
print(p_value)
print(dof)
print(expected)
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("rejected")
else:
    print("accepted")
```
